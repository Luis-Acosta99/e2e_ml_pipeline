```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10, 6))

# 1. Plot Condition 0 on the primary (left) Y-axis
color0 = 'tab:blue'
df_train[df_train['anomaly'] == 0]['meter_reading'].hist(
    range=[0, 1000], bins=50, ax=ax1, alpha=0.5, color=color0, label='Condition 0'
)
ax1.set_xlabel('Meter Reading')
ax1.set_ylabel('Normal', color=color0)
ax1.tick_params(axis='y', labelcolor=color0)

# 2. Create a twin axis for Condition 1
ax2 = ax1.twinx()

# 3. Plot Condition 1 on the secondary (right) Y-axis
color1 = 'tab:orange'
df_train[df_train['anomaly'] == 1]['meter_reading'].hist(
    range=[0, 1000], bins=50, ax=ax2, alpha=0.5, color=color1, label='Condition 1'
)
ax2.set_ylabel('Anomaly', color=color1)
ax2.tick_params(axis='y', labelcolor=color1)

# 4. Final touches
plt.title('Overlaid Histograms with Independent Y-Scales')
ax1.grid(False) 
ax2.grid(False)
plt.show()
```

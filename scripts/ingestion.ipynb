```python
import pandas as pd
import os
from pathlib import Path
```

```python
input_path = "../data/energy-anomaly-detection/"
output_path = "../data/bronze/"
```

```python
def convert_csv_to_parquet(input_folder, output_folder):
    Path(output_folder).mkdir(parents=True, exist_ok=True)
    
    for filename in os.listdir(input_folder):
        if filename.endswith('.csv'):
            csv_path = os.path.join(input_folder, filename)
            
            parquet_filename = filename.replace('.csv', '.parquet')
            parquet_path = os.path.join(output_folder, parquet_filename)
            
            print(f"Converting {filename}...")
            
            try:
                df = pd.read_csv(csv_path,header=0, low_memory=False)
                df.to_parquet(parquet_path, engine='pyarrow', index=False)
                print(f"Successfully saved to: {parquet_path}")
            except Exception as e:
                print(f"Error converting {filename}: {e}")
```

```python
convert_csv_to_parquet(input_path, output_path)
```
